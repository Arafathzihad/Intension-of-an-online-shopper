```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['AWARD_DATE']
  ys = series['QUANTITY']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = data.sort_values('AWARD_DATE', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('UNIT_OF_MEASURE')):
  _plot_series(series, series_name, i)
  fig.legend(title='UNIT_OF_MEASURE', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('AWARD_DATE')
_ = plt.ylabel('QUANTITY')
```

```python
import pandas as pd

def find_duplicate_columns(data):

    duplicate_columns = {}


    for i in range(data.shape[1]):

        col1 = data.iloc[:, i]
        for j in range(i + 1, data.shape[1]):

            col2 = data.iloc[:, j]

            if col1.equals(col2):

                col1_name = data.columns[i]
                col2_name = data.columns[j]

                if col1_name not in duplicate_columns:
                    duplicate_columns[col1_name] = [col2_name]
                else:
                    duplicate_columns[col1_name].append(col2_name)

    return duplicate_columns
```

```python
duplicates = find_duplicate_columns(data)
print(duplicates)
```

```python
from sklearn import preprocessing


for col in data.select_dtypes(include=['object','bool']).columns:


    print(f"{col}: {data[col].unique()}")


for col in data.select_dtypes(include=['object','bool']).columns:

    data[col].fillna('unknown', inplace=True)


    data[col] = data[col].astype(str)


    label_encoder = preprocessing.LabelEncoder()


    label_encoder.fit(data[col].unique())


    data[col] = label_encoder.transform(data[col])


    print(f"{col}: {data[col].unique()}")
```

```python
X = data.drop(['COMMODITY'], axis=1)
y = data['COMMODITY']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)  # You can change the number of neighbors

# Fit the classifier to the data
knn.fit(X_train, y_train)

# Predict on test data
y_pred_knn = knn.predict(X_test)

# Evaluate the model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"Accuracy of KNN: {accuracy_knn}")
print(classification_report(y_test, y_pred_knn))
```

```python
# Create a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)  # You can modify the number of trees

# Fit the model on the training data
rf.fit(X_train, y_train)

# Predict on test data
y_pred_rf = rf.predict(X_test)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Accuracy of Random Forest: {accuracy_rf}")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# For K-Nearest Neighbors Regression
knn_reg = KNeighborsRegressor(n_neighbors=3)
knn_reg.fit(X_train, y_train)
y_pred_knn_reg = knn_reg.predict(X_test)
mse_knn = mean_squared_error(y_test, y_pred_knn_reg)
print(f"Mean Squared Error for KNN Regressor: {mse_knn}")
print(f"R^2 Score for KNN Regressor: {r2_score(y_test, y_pred_knn_reg)}")

# For Random Forest Regression
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train, y_train)
y_pred_rf_reg = rf_reg.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf_reg)
print(f"Mean Squared Error for Random Forest Regressor: {mse_rf}")
print(f"R^2 Score for Random Forest Regressor: {r2_score(y_test, y_pred_rf_reg)}")
```
